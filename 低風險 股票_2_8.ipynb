```python
#新增現金股利id
primary_key_現金股利 = []
for i in range(len(df_現金股利['盈餘分派_年度'])) :
    primary_key_現金股利.append(f'{df_現金股利["公司"].values[i]}_{df_現金股利["盈餘分派_年度"].values[i]}')
    #print(df_現金股利.index.values[i],df_現金股利['盈餘分派_年度'].values[i])
df_現金股利['id'] = primary_key_現金股利

#新增每股稅後淨利id再把12月以外的資料刪除
print('df_每股稅後淨利',len(df_每股稅後淨利['公司']))
primary_key_每股稅後淨利 = []
drop_index = []
for i in range(len(df_每股稅後淨利['公司']))   :
    if str(df_每股稅後淨利["年/月"].values[i])[-2:] == '12' :
        primary_key_每股稅後淨利.append(f'{df_每股稅後淨利["公司"].values[i]}_{str(df_每股稅後淨利["年/月"].values[i])[:-2]}')
    else :  
        drop_index.append(i)       
df_每股稅後淨利 = df_每股稅後淨利.drop(df_每股稅後淨利.index[drop_index])
df_每股稅後淨利['id'] = primary_key_每股稅後淨利 

#刪除 df_每股稅後淨利 和 df_現金股利 無id交集的資料
print('droped df_每股稅後淨利',len(df_每股稅後淨利['公司']))   
drop_index = []
for i in df_每股稅後淨利.index :
    if df_每股稅後淨利['id'][i] not in df_現金股利['id'].values  :  
        drop_index.append(i)
df_每股稅後淨利 = df_每股稅後淨利.drop(drop_index)

drop_index = []
for i in df_現金股利.index :
    if df_現金股利['id'][i] not in df_每股稅後淨利['id'].values  :  
        drop_index.append(i)
df_現金股利 = df_現金股利.drop(drop_index)
print('droped df_每股稅後淨利',len(df_每股稅後淨利['id']))   
print('df_每股稅後淨利',len(df_現金股利['id']))

#merge by id 
df_每股稅後淨利 = pd.DataFrame(list(zip(df_每股稅後淨利['id'],df_每股稅後淨利['每股盈餘'])),columns = ['id','每股盈餘'])
df_現金股利 = pd.DataFrame(pd.merge(df_現金股利, df_每股稅後淨利, on=['id']),
    columns=['id','簡稱', '現金股利(元)_盈餘','每股盈餘'])

#df_現金股利['每股盈餘']補0
for i in range(len(df_現金股利['每股盈餘'])) :
    if type(df_現金股利['每股盈餘'][i]) == str :
        print(df_現金股利['每股盈餘'][i])
        df_現金股利['每股盈餘'][i] = int(0)
```

```python
all_company = list(set(list(df_現金股利['簡稱'])))
盈餘_list = []
每股盈餘_list = []
std_list = []
company_num_list = []
for i in all_company :
    mean = 0
    each_mean = 0 
    c = 0 
    _std_list = []
    for j in range(len(df_現金股利['id'])) :
        if i == df_現金股利['簡稱'][j] :
            company_num_list.append(int(df_現金股利['id'][j][:-5]))
            mean += df_現金股利['現金股利(元)_盈餘'][j]
            each_mean += df_現金股利['每股盈餘'][j]
            c += 1
            _std_list.append(df_現金股利['現金股利(元)_盈餘'][j])
    if c != 0:
        mean /= c
        each_mean /= c
    std_list.append(np.std(np.array(_std_list), ddof = 1))    
    每股盈餘_list.append(each_mean)
    盈餘_list.append(mean)
company_num_list = []
for i in all_company :
    for j in range(len(df_現金股利['id'])) :
        if i == df_現金股利['簡稱'][j] :
            company_num_list.append(int(df_現金股利['id'][j][:-5])) 
            break 
sharp_list = np.array(盈餘_list)/np.array(std_list)
pd_target = pd.DataFrame(list(zip(company_num_list,all_company,盈餘_list,std_list,sharp_list,每股盈餘_list)),
    columns = ['公司','簡稱','盈餘mean','現金股利(元)_盈餘std','現金股利(元)_盈餘sharp','每股盈餘mean'])
pd_target = pd_target.sort_values(by = '公司')     
pd_target       
```

```python
df_未調整=data.get('未調整的 股價+報酬率')
df_未調整_收= pd.DataFrame(df_未調整,columns=['證券代碼','年月日','簡稱','開盤價','收盤價'])
df_未調整_收['股價日報酬率'] = (df_未調整_收['收盤價']-df_未調整_收['開盤價'])/df_未調整_收['開盤價']
df_未調整_收['年月日'] = df_未調整_收['年月日'].apply(str)
df_未調整_收
```

```python
def get_mean_stock(stock_num) :
    list_1101 = []
    c_2014 = []
    c_2015 = []
    c_2016 = []
    c_2017 = []
    c_2018 = []
    c_2019 = []
    for i in range(len(df_未調整_收[df_未調整_收['證券代碼'] == stock_num])) :        
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2014' :
            c_2014.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i])
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2015' :
            c_2015.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i])
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2016' :
            c_2016.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i])
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2017' :
            c_2017.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i]) 
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2018' :
            c_2018.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i])
        if df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['年月日'].values[i][:4] == '2019' :
            c_2019.append(df_未調整_收[df_未調整_收['證券代碼'] == stock_num]['股價日報酬率'].values[i]) 
    list_1101.append(1000*np.mean(c_2014))
    list_1101.append(1000*np.mean(c_2015)) 
    list_1101.append(1000*np.mean(c_2016))
    list_1101.append(1000*np.mean(c_2017))
    list_1101.append(1000*np.mean(c_2018))
    list_1101.append(1000*np.mean(c_2019))
    list_1101.append(np.std(c_2014))
    list_1101.append(np.std(c_2015)) 
    list_1101.append(np.std(c_2016))
    list_1101.append(np.std(c_2017))
    list_1101.append(np.std(c_2018))
    list_1101.append(np.std(c_2019))    
    return list_1101
```

```python
import time
list_mean_2014 = []
list_mean_2015 = []
list_mean_2016 = []
list_mean_2017 = []
list_mean_2018 = []
list_mean_2019 = []
list_std_2014 = []
list_std_2015 = []
list_std_2016 = []
list_std_2017 = []
list_std_2018 = []
list_std_2019 = []
for i in all_簡稱 :
    ticks = time.time()
    _list = get_mean_stock(i)
    print(i,time.time()-ticks)
    list_mean_2014.append(_list[0])
    list_mean_2015.append(_list[1])
    list_mean_2016.append(_list[2])
    list_mean_2017.append(_list[3])
    list_mean_2018.append(_list[4])
    list_mean_2019.append(_list[5])
    list_std_2014.append(_list[6])
    list_std_2015.append(_list[7])
    list_std_2016.append(_list[8])
    list_std_2017.append(_list[9])
    list_std_2018.append(_list[10])
    list_std_2019.append(_list[11])
pd_avg = pd.DataFrame(list(zip(all_簡稱[:5],list_mean_2014,list_mean_2015,list_mean_2016,list_mean_2017,list_mean_2018,list_mean_2019,list_std_2014,list_std_2015,list_std_2016,list_std_2017,list_std_2018,list_std_2019)),
    columns = ['公司','2014_mean','2015_mean','2016_mean','2017_mean','2018_mean','2019_mean','2014_std','2015_std',           '2016_std','2017_std','2018_std','2019_std'])
pd_avg    
```
